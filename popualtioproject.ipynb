```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Load in dataset
populations = pd.read_csv('population_by_country_2020.csv')
```

```python
# Made our DataFrame
df = pd.DataFrame(populations)
```

```python
# Explore the dataset
df.head(10)
```

```python
#Fill in all the N.A entries with 0.
df.replace(np.nan,0)
```

```python
df
```

```python
# Select the Classifier
from sklearn.linear_model import LinearRegression
```

```python
#Select the Model 
from sklearn.model_selection import train_test_split
```

```python
# Features we're starting off with since other features contain NaN
x = df[['Land Area (Km²)', 'Density (P/Km²)']]
```

```python
# Target
y = df[['Population (2020)']]
```

```python
# TTS lets us split our data into easily processed chunks
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)
```

```python
#Fit the model
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("MAE: {}".format(np.abs(y_test-y_pred).mean()))
print("RMSE: {}".format(np.sqrt(((y_test-y_pred)**2).mean())))
```

```python
#Plotting the figure
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(6, 4))
plt.clf()

elev = 43.5
azim = -110

ax = Axes3D(fig, elev=elev, azim=azim)


ax.set_xlabel('Land Area (Km²)')
ax.set_ylabel('Density (P/Km²)')
ax.set_zlabel('Population')

#Scatter Plot
ax.scatter(x_train[['Land Area (Km²)']], x_train[['Density (P/Km²)']], y_train, c='k', marker='+')
ax.plot_surface(np.array([[0, 0], [16376870, 16376870]]), np.array([[0, 26337], [0, 26337]]), model.predict(np.array([[0, 0, 16376870, 16376870], [0, 26337, 0, 26337]]).T).reshape((2, 2)), alpha=.7)

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
